```python
import random   # Importing Library

                # 3 x 3 Board
board = [[' ' for _ in range(3)] for _ in range(3)]

                # 2 Players
                # Human -> X
                # Computer -> O
human = 'X'
computer = 'O'

                        # MinMax algorithm
def MinMax(board, player):
    
    if Check_Win(board, human):
        return -1
    elif Check_Win(board, computer):
        return 1
    elif Check_Draw(board):
        return 0
    
    if player == computer:    
        best_score = -2    # For the maximizing player (computer)
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = computer
                    score = MinMax(board, human)    # Now human turn
                    board[i][j] = ' '
                    best_score = max(score, best_score)
        return best_score
    
    else:
        best_score = 2      # For the minimizing player (human)
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = human
                    score = MinMax(board, computer)    # Now computer turn
                    board[i][j] = ' '
                    best_score = min(score, best_score)
        return best_score


                    # Defining the winning terminal states
def Check_Win(board, player):
    for i in range(3):
        if board[i][0] == player and board[i][1] == player and board[i][2] == player:
            return True
        if board[0][i] == player and board[1][i] == player and board[2][i] == player:
            return True
    if board[0][0] == player and board[1][1] == player and board[2][2] == player:
        return True
    if board[0][2] == player and board[1][1] == player and board[2][0] == player:
        return True
    return False

                    # Defining the draw (tie) terminal states
def Check_Draw(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                return False
    return True


                    # Function to print the Board
def Print_Board(b):
    print("    1   2   3 ")
    print("1-> " + b[0][0] + " | " + b[0][1] + " | " + b[0][2] + " ")
    print("   ---|---|---")
    print("2-> " + b[1][0] + " | " + b[1][1] + " | " + b[1][2] + " ")
    print("   ---|---|---")
    print("3-> " + b[2][0] + " | " + b[2][1] + " | " + b[2][2] + " ")

                        # Computer Moves
def Computer_Move(board):
    best_score = -2
    best_move = []  # Stores row and column indices that represents the best move for the computer 
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = computer
                score = MinMax(board, human)   # Calling MinMax func -> human
                board[i][j] = ' '
                if score > best_score:   # Checks if the calculated score is better than the previous best score, 
                                        # Updates best_score and best_move if a better score is found
                    best_score = score
                    best_move = [i, j]
    board[best_move[0]][best_move[1]] = computer

    
def TicTacToe():   # Main game func
    
    print("    *-*-*-Welcome to Tic Tac Toe!-*-*-*\n")
    
    Print_Board(board)   # Printing the Board
    
    while not Check_Win(board, human) and not Check_Win(board, computer) and not Check_Draw(board):
                    # loop will continue to execute as long as none of these three conditions are true.
        
        # Your (human) turn
        print("\nIt's Your turn.")
#         row = int(input("Enter the row number (1-3): "))
#         col = int(input("Enter the column number (1-3): "))
                
                        # Give input with comma separated
        try:
            row, col = input('Enter your move (row, column)[1 - 3]: ').split(',')
            row = int(row.strip())
            col = int(col.strip())
            
            if not (1 <= row <= 3) or not (1 <= col <= 3):   # When user enters value higher than 3
                raise ValueError("Row and column values must be between 1 and 3.")

        except ValueError as e:
            print('Invalid input:', e)  # Displays the error message
            continue
        except IndexError:
            print('Invalid move!')
            continue
            
        if board[row - 1][col - 1] != ' ':
            print("Invalid move, try again.")   # Already occupied
            continue
            
        board[row - 1][col - 1] = human
        Print_Board(board)      # Prints updated board
        
        if Check_Win(board, human):   # Checking win terminal states
            print("\nPlayer X Wins! (You Win)")
            return
        
        elif Check_Draw(board):    # Checking draw terminal states
            print("\nIt's a DRAW (Tie)!")
            return

               # Computer's turn
            
        print("\nIt's Computer turn.")
        Computer_Move(board)
        Print_Board(board)     # Prints updated board
        
        if Check_Win(board, computer):   # Checking win terminal states
            print("\nPlayer O Wins! (Computer Wins)")
            return
        
        elif Check_Draw(board):     # Checking draw terminal states
            print("\nIt's a DRAW (Tie)!")
            return
        
TicTacToe()   # Calling the function
```
